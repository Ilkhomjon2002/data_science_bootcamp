```python
import seaborn as  sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False)
plt.show()
```

```python
sns.distplot(tips['total_bill'])
plt.show()
```

```python
sns.distplot(tips['total_bill'], kde=False,bins=30)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde',hue='sex')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
plt.show()
```

```python
sns.rugplot(tips['total_bill'])
plt.show()
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill'  ,data=tips,estimator='max') #default estimator is mean
plt.show()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill'  ,data=tips,estimator=np.std) #default estimator is mean
plt.show()
```

```python
sns.countplot(x='sex',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',jitter=True)
plt.show()
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
sns.violinplot(x='day',y='total_bill',data=tips)
plt.show()
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
tips.head()
```

```python
fp =flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
plt.show()
```

```python
fp =flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fp,cmap='coolwarm',linecolor='black',linewidths=1)
plt.show()
```

```python
fp =flights.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fp)
plt.show()
```

```python
sns.clustermap(fp)
plt.show()
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
plt.show()
```

```python
g = sns.FacetGrid(data=tips,col='time', row='smoker')
g.map(plt.scatter,'total_bill','tip')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':50})
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',aspect=0.6)
plt.show()
```
