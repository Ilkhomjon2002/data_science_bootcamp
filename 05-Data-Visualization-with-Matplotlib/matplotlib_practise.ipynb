```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
# functional method
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
plt.show()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(5,5))
#linewidth is same as lw
#linestyle is same as ls
axes.plot(x,x*1,label='x*1',color='r', linewidth=1, marker='1')
axes.plot(x,x**2,label='x**2',color ='b',lw=2,linestyle = '-.',marker= '+',markersize=10)
axes.plot(x,x**3,label='x**3',color='#FF8C00',linewidth=3,alpha=0.8,marker='o',markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='g')
axes.legend(loc=0)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw='2',ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
plt.show()
```
